In this file, the goal is to take a day's worth of data, and calculate the cumulative distance traveled by a single scooter. Since we are given a lat long, the plan is to group by a scooters ID, calculate cumulative distance, and then ungroup the data.

```python
import pandas as pd
import geopandas as gpd
```

```python
Scooters_2023_05_11 = pd.read_csv('Data/Scooters 2023-05-11.csv', index_col=0)
Scooters_2023_05_11.head()
```

```python
grouped_scooters = Scooters_2023_05_11.groupby('id')[['latitude','longitude']].count()
```
